```python
from hangul_utils import split_syllables, join_jamos
from khaiii import KhaiiiApi
from collections import deque
api = KhaiiiApi()

da_low= ['다', 'ㄴ다', '는다', '느냐다', '단다', '서다', '란다', 'ㄴ단다'] 

def Changer(input):

    result = []

    analyzed = api.analyze(input)  
    for data in analyzed:
        lis_word = []
        lis_tag = []
        EF_in = 0
        for morph in data.morphs:
            lis_word.append(morph.lex)
            lis_tag.append(morph.tag)
            if morph.tag == 'EF':
                EF_in = 1
        if EF_in == 1:
            result.append(to_high(lis_word,lis_tag))
        else:
            result.append(data.lex)

    return ' '.join(result)
        
def to_high(lis_word, lis_tag):
#     print(lis_word)
#     print(lis_tag)
    result = ''
    for i in range(len(lis_word)):
        #다로 끝나는 애들
        if lis_tag[i] == 'EF' and lis_word[i] in da_low:
            jamo = split_syllables(lis_word[i-1][-1])[-1]
            if lis_word[i] == '다':
                #받침있으면 습니다
                if jamo >= 'ㄱ' and jamo <= 'ㅎ':
                    lis_word[i] = '습니다'
                #받침없으면 ㅂ니다
                else:
                    lis_word[i] = 'ㅂ니다'
            elif lis_word[i] == '는다':    
                lis_word[i] = '습니다'
            elif lis_word[i] == 'ㄴ다':    
                lis_word[i] = 'ㅂ니다'
            elif lis_word[i] == '느냐다':    
                lis_word[i] = '느냐입니다'
            elif lis_word[i] == '서다':    
                lis_word[i] = '서입니다'
            elif lis_word[i] == '단다':    
                lis_word[i] = '답니다'
            elif lis_word[i] == '란다':    
                lis_word[i] = '랍니다'
            elif lis_word[i] == 'ㄴ단다':    
                lis_word[i] = 'ㄴ답니다'
        
        if lis_word[i][0] >= 'ㄱ' and lis_word[i][0] <= 'ㅎ' :
            if lis_word[i-1][-1] >= 'ㄱ' and lis_word[i-1][-1] <= 'ㅎ' :
                result = result[:-1] + join_jamos(split_syllables(result[-1][:-1])+split_syllables(lis_word[i]))
            else:
                result = result[:-1] + join_jamos(split_syllables(result[-1])+split_syllables(lis_word[i]))
        else:
            result+= lis_word[i]
    return result
```
