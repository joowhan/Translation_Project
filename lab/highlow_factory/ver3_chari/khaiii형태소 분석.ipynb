```python
#khaiii
from tqdm import tqdm
from khaiii import KhaiiiApi
api = KhaiiiApi()

###khaiii으로 나눠서 파일 다운로드하기
opend = open("/Users/kimjoowhan/Documents/University/Capston Project/aihub.kr", "r", encoding = "utf-8")

lis_word_file = open("/Users/kimjoowhan/Documents/University/Capston Project/lis/lis_word", "w", encoding = "utf-8")
lis_tag_file = open("/Users/kimjoowhan/Documents/University/Capston Project/lis/lis_tag", "w", encoding = "utf-8")

for i in tqdm(range(3525728)):
    word = ''
    tag = ''
    read = opend.readline()
    if read[-1]=='\n':
        read = read.replace('\n','')
    if read[-1] =='?' or read[-1] =='.' or read[-1] =='!' or read[-1] =='\"':
        read = read
    else:
        read = read+'.'
    analyzed = api.analyze(read)    
    for test in analyzed:
        for morph in test.morphs:
            word += morph.lex + ' '
            tag += morph.tag + ' '
    lis_word_file.write(word)
    lis_tag_file.write(tag)
    if (i != 1600000-1):
        lis_word_file.write('\n')
        lis_tag_file.write('\n')
            
opend.close()
lis_word_file.close()
lis_tag_file.close()
```

```python
from khaiii import KhaiiiApi
api = KhaiiiApi()

##Use in Changer
def to2lists(input):
    lis_word = []
    lis_tag = []
    analyzed = api.analyze(input)  
    for data in analyzed:
        for morph in data.morphs:
            lis_word.append(morph.lex)
            lis_tag.append(morph.tag)
    return lis_word, lis_tag
```
